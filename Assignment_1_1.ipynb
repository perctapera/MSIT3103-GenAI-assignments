```python
!pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu121
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
import os

# Set device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
```

```python
def load_dataset(dataset_name='MNIST', batch_size=64, image_size=28):
    transform = transforms.Compose([
        transforms.Resize(image_size),
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))  # Normalize to [-1, 1]
    ])
    
    if dataset_name == 'MNIST':
        dataset = torchvision.datasets.MNIST(root='./data', train=True, download=True, transform=transform)
        channels = 1
    elif dataset_name == 'Fashion-MNIST':
        dataset = torchvision.datasets.FashionMNIST(root='./data', train=True, download=True, transform=transform)
        channels = 1
    elif dataset_name == 'CIFAR-10':
        dataset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
        channels = 3
        # For CIFAR, normalize properly if needed, but keeping simple
    else:
        raise ValueError("Unsupported dataset")
    
    dataloader = torch.utils.data.DataLoader(dataset, batch_size=batch_size, shuffle=True)
    return dataloader, channels

# Example: Load MNIST
dataloader, channels = load_dataset('Fashion-MNIST', batch_size=64)
print(f"Dataset loaded: channels={channels}, batch_size=64")
```

```python
def visualize_samples(dataloader, num_samples=16):
    dataiter = iter(dataloader)
    images, _ = next(dataiter)
    images = images[:num_samples]
    
    fig, axes = plt.subplots(4, 4, figsize=(8, 8))
    for i, ax in enumerate(axes.flat):
        img = images[i].squeeze() if channels == 1 else images[i].permute(1, 2, 0)
        img = (img + 1) / 2  # Denormalize to [0,1]
        ax.imshow(img, cmap='gray' if channels == 1 else None)
        ax.axis('off')
    plt.tight_layout()
    plt.show()

visualize_samples(dataloader)
```

```python
class Generator(nn.Module):
    def __init__(self, latent_dim=100, img_channels=1, img_size=28):
        super(Generator, self).__init__()
        self.latent_dim = latent_dim
        self.img_channels = img_channels
        self.img_size = img_size
        
        # Simple FC-based for MNIST/Fashion (28x28=784)
        self.model = nn.Sequential(
            nn.Linear(latent_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 256),
            nn.ReLU(),
            nn.Linear(256, 512),
            nn.ReLU(),
            nn.Linear(512, img_size * img_size * img_channels),
            nn.Tanh()  # Output to [-1,1]
        )
    
    def forward(self, z):
        img = self.model(z)
        img = img.view(img.size(0), self.img_channels, self.img_size, self.img_size)
        return img

# For CIFAR-10, you'd use a DCGAN-style with ConvTranspose2d; extend here if needed
# Example instantiation
latent_dim = 100
generator = Generator(latent_dim, channels).to(device)
print(generator)
```

```python
class Discriminator(nn.Module):
    def __init__(self, img_channels=1, img_size=28):
        super(Discriminator, self).__init__()
        self.img_channels = img_channels
        self.img_size = img_size
        
        # Flatten input: img_size * img_size * channels
        self.model = nn.Sequential(
            nn.Linear(img_size * img_size * img_channels, 512),
            nn.LeakyReLU(0.2),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2),
            nn.Linear(256, 128),
            nn.LeakyReLU(0.2),
            nn.Linear(128, 1),
            nn.Sigmoid()  # Probability
        )
    
    def forward(self, img):
        img_flat = img.view(img.size(0), -1)
        validity = self.model(img_flat)
        return validity

discriminator = Discriminator(channels).to(device)
print(discriminator)
```

```python
def weights_init(m):
    if isinstance(m, nn.Linear):
        torch.nn.init.xavier_uniform_(m.weight)
        m.bias.data.fill_(0.01)

generator.apply(weights_init)
discriminator.apply(weights_init)
```

```python
criterion = nn.BCELoss()

lr_g = 0.0002  # Generator learning rate
lr_d = 0.0002  # Discriminator learning rate
optimizer_g = optim.Adam(generator.parameters(), lr=lr_g, betas=(0.5, 0.999))
optimizer_d = optim.Adam(discriminator.parameters(), lr=lr_d, betas=(0.5, 0.999))
```

```python
def train_gan(dataloader, generator, discriminator, num_epochs=50, latent_dim=100, save_dir='results'):
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    
    criterion = nn.BCELoss()
    optimizer_g = optim.Adam(generator.parameters(), lr=0.0002, betas=(0.5, 0.999))
    optimizer_d = optim.Adam(discriminator.parameters(), lr=0.0002, betas=(0.5, 0.999))
    
    g_losses, d_losses = [], []
    
    for epoch in range(num_epochs):
        for i, (real_imgs, _) in enumerate(tqdm(dataloader)):
            batch_size = real_imgs.size(0)
            real_imgs = real_imgs.to(device)
            
            # Labels
            real_label = torch.ones(batch_size, 1).to(device)
            fake_label = torch.zeros(batch_size, 1).to(device)
            
            # Train Discriminator
            optimizer_d.zero_grad()
            real_output = discriminator(real_imgs)
            d_real_loss = criterion(real_output, real_label)
            
            z = torch.randn(batch_size, latent_dim).to(device)
            fake_imgs = generator(z)
            fake_output = discriminator(fake_imgs.detach())
            d_fake_loss = criterion(fake_output, fake_label)
            
            d_loss = d_real_loss + d_fake_loss
            d_loss.backward()
            optimizer_d.step()
            
            # Train Generator
            optimizer_g.zero_grad()
            fake_output = discriminator(fake_imgs)
            g_loss = criterion(fake_output, real_label)  # Fool D
            g_loss.backward()
            optimizer_g.step()
            
            g_losses.append(g_loss.item())
            d_losses.append(d_loss.item())
        
        print(f"Epoch [{epoch+1}/{num_epochs}] D_loss: {d_losses[-1]:.4f} G_loss: {g_losses[-1]:.4f}")
        
        # Save sample images every 10 epochs
        if (epoch + 1) % 10 == 0:
            with torch.no_grad():
                z = torch.randn(16, latent_dim).to(device)
                fake_imgs = generator(z).cpu()
                visualize_samples_from_tensor(fake_imgs)
                plt.savefig(f'{save_dir}/epoch_{epoch+1}.png')
    
    return g_losses, d_losses

# Helper to visualize generated samples
def visualize_samples_from_tensor(images, num_samples=16):
    fig, axes = plt.subplots(4, 4, figsize=(8, 8))
    for i, ax in enumerate(axes.flat):
        img = images[i].squeeze() if channels == 1 else images[i].permute(1, 2, 0)
        img = (img + 1) / 2
        ax.imshow(img, cmap='gray' if channels == 1 else None)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
# Baseline
num_epochs = 50
g_losses, d_losses = train_gan(dataloader, generator, discriminator, num_epochs, latent_dim)
plt.plot(g_losses, label='G Loss')
plt.plot(d_losses, label='D Loss')
plt.legend()
plt.show()
```

```python
# Define experiments
experiments = {
    'baseline': {'batch_size': 64, 'lr_g': 0.0002, 'lr_d': 0.0002, 'latent_dim': 100},
    'high_lr': {'batch_size': 64, 'lr_g': 0.0004, 'lr_d': 0.0004, 'latent_dim': 100},  # Higher LR: may diverge
    'low_lr': {'batch_size': 64, 'lr_g': 0.0001, 'lr_d': 0.0001, 'latent_dim': 100},   # Slower convergence
    'large_batch': {'batch_size': 128, 'lr_g': 0.0002, 'lr_d': 0.0002, 'latent_dim': 100},  # Larger batch: smoother but slower?
    'small_latent': {'batch_size': 64, 'lr_g': 0.0002, 'lr_d': 0.0002, 'latent_dim': 64},    # Smaller noise: less variety?
    'large_latent': {'batch_size': 64, 'lr_g': 0.0002, 'lr_d': 0.0002, 'latent_dim': 128}    # More capacity
}

# Modified train function to accept hypers (simplified; in practice, recreate optimizers inside)
def train_with_hypers(exp_name, hypers, num_epochs=20):  # Shorter for multiple runs
    batch_size = hypers['batch_size']
    dataloader_exp, _ = load_dataset('MNIST', batch_size)  # Reload for batch_size
    latent_dim = hypers['latent_dim']
    generator_exp = Generator(latent_dim, channels).to(device).apply(weights_init)
    discriminator_exp = Discriminator(channels).to(device).apply(weights_init)
    
    # Custom optimizers with hypers
    optimizer_g = optim.Adam(generator_exp.parameters(), lr=hypers['lr_g'], betas=(0.5, 0.999))
    optimizer_d = optim.Adam(discriminator_exp.parameters(), lr=hypers['lr_d'], betas=(0.5, 0.999))
    
    # Train (reuse logic, but update optimizer in train_gan if needed; for simplicity, inline key parts)
    g_losses, d_losses = [], []
    for epoch in range(num_epochs):
        for i, (real_imgs, _) in enumerate(tqdm(dataloader_exp)):
            batch_size_i = real_imgs.size(0)
            real_imgs = real_imgs.to(device)
            real_label = torch.ones(batch_size_i, 1).to(device)
            fake_label = torch.zeros(batch_size_i, 1).to(device)
            
            # D train
            optimizer_d.zero_grad()
            real_output = discriminator_exp(real_imgs)
            d_real_loss = criterion(real_output, real_label)
            z = torch.randn(batch_size_i, latent_dim).to(device)
            fake_imgs = generator_exp(z)
            fake_output = discriminator_exp(fake_imgs.detach())
            d_fake_loss = criterion(fake_output, fake_label)
            d_loss = (d_real_loss + d_fake_loss) / 2  # Average
            d_loss.backward()
            optimizer_d.step()
            
            # G train
            optimizer_g.zero_grad()
            fake_output = discriminator_exp(fake_imgs)
            g_loss = criterion(fake_output, real_label)
            g_loss.backward()
            optimizer_g.step()
            
            g_losses.append(g_loss.item())
            d_losses.append(d_loss.item())
        
        if (epoch + 1) % 10 == 0:
            print(f"{exp_name} Epoch [{epoch+1}/{num_epochs}] D: {d_losses[-1]:.4f} G: {g_losses[-1]:.4f}")
            # Generate sample
            with torch.no_grad():
                z_sample = torch.randn(16, latent_dim).to(device)
                fake_sample = generator_exp(z_sample).cpu()
                plt.figure()
                visualize_samples_from_tensor(fake_sample)
                plt.title(f"{exp_name} - Epoch {epoch+1}")
                plt.savefig(f'results/{exp_name}_epoch_{epoch+1}.png')
    
    # Final plot
    plt.figure()
    plt.plot(g_losses, label='G Loss')
    plt.plot(d_losses, label='D Loss')
    plt.title(f"{exp_name} Losses")
    plt.legend()
    plt.savefig(f'results/{exp_name}_losses.png')
    plt.show()
    
    # Save final generator for later comparison
    torch.save(generator_exp.state_dict(), f'results/{exp_name}_generator.pth')
    
    return g_losses, d_losses

# Run experiments (this will take time; run selectively)
results = {}
for name, hypers in experiments.items():
    print(f"\nRunning {name}...")
    results[name] = train_with_hypers(name, hypers, num_epochs=20)
```
